```python
import pandas as pd
```

```python
filename1 = 'Killed_Targets-'
filename2 = 1
filename3 = '.csv'

i = 1
j = 1
kt_list = []
while i < 11:
    if j <= 1:
        kt_list.append(str(i) + ' kills')
        j += 1
    else:
        j = 1
        kt_list.append(str(i) + ' tracking')
        i += 1

collist = ['num_t','num_m','num_s','seed', 'max_samp','speed_s','P_k','Lam_T','Lam_M','speed_t']
test_jobs = pd.read_csv('test-jobz.csv', header = None)
test_jobs.drop(0,axis =1, inplace = True)
test_jobs.set_axis(collist,inplace = True, axis = 1)

zeros = [0]*len(test_jobs)
for item in kt_list:
    test_jobs[item] = zeros

while filename2 < 1089:
    filename = filename1 + str(filename2) + filename3
    Killed_Targets = pd.read_csv(filename)
    Killed_Targets.drop('Unnamed: 0', axis =1, inplace = True)
    Killed_Targets.fillna(0, inplace = True)
    try:
        Killed_Targets.apply(pd.to_numeric)
    except:
        print(filename2)
    for item in kt_list:
        test_jobs.loc[(filename2 - 1),item] = float(Killed_Targets[item].mean())
    
    filename2 += 1
    
```
